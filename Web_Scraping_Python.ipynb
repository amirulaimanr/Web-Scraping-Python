```python
news_url = "https://www.thestar.com.my/news/nation/2023/03/30/couple-unaware-of-toxins"
```

```python
# Import BeautifulSoup Library
from bs4 import BeautifulSoup

# Import library to consume URL
import requests

# Import libraary for regular expression to extract information
import re
```

```python
# Consume the URL to get the content ( HTML )
response = requests.get(news_url)
print(f"Response from requests: {response}")
```

```python
print(response.content)
```

```python
content = BeautifulSoup(response.content, "html.parser")
print(content)
```

```python
# Search by HTML TAG ( Element )
title_options = content.find_all("h1")
print(f"Data Type: {type(title_options)}")
print(f"Title: {title_options[0]}")
print(f"Extract value from H1 : {title_options[0].text}")
print(f"Clean value : {title_options[0].text.strip()}")
title = title_options[0].text.strip()
print(f"Cleaned Title : {title}")
```

```python
# Extract Image with class name
image_options = content.find_all("div", {"class": "story-image"})
print(f"Step 1 : Image Options 1st value : {image_options[0]}")

print(f"Step 2 : Get Image Script value : {image_options[0].script}")

print(f"Step 3 : Content of the Script : {image_options[0].script.text}")

# Regular Expression ( https://regex101.com )
generate_media_tag_pattern = re.compile(r'GenerateMediaTag\((\{[\w\W]+\})')

print(f"Step 4 : Prepare regular expression to extract content")
script_value = image_options[0].script.text

print(f"Step 5 : Assign script value into a variable : {script_value}")
matched_content = generate_media_tag_pattern.findall(script_value)

print(f"Step 6 : Matched Content : {matched_content}")

print(f"Step 7 : Data Type for matched content {type(matched_content)}")

print(f"Step 8 : JSON String Content : {matched_content[0]}")

print(f"Import json library")

import json
generate_media_tag_dict = json.loads(matched_content[0])

print(f"Step 9 : Convert JSON String into Dictionary Format : {generate_media_tag_dict}")

print(f"Step 9 : Data Type : {type(generate_media_tag_dict)}")

print(f"Step 10 : Image path : {generate_media_tag_dict['image_path']})")

print(f"Step 11 : Full image path : https://www.thestar.com.my/{generate_media_tag_dict['image_path']})")

image_path = f"https://www.thestar.com.my/{generate_media_tag_dict['image_path']}"
```
